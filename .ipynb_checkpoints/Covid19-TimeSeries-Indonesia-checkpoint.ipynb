# Split Data

```python
def process_data(data):
    x,y = [],[]
    switch = False
    
    if len(data)%2 == 1:
        last = False
    else:
        last = True
    
    for i in range(len(data)):
        if i == len(data)-1:
            if not last:
                break
        if switch:
            y.append(data[i])
            switch = False
        else:
            x.append(data[i])
            switch = True
    
    def reshape(d):
        d = np.array(d)
        d = np.reshape(d,(d.shape[0],1,1))
        return d
    return (reshape(x),np.array(y))
```

```python
death = dataframe['Total Deaths']
```

```python
recovered = dataframe['Total Recovered']
```

```python
death_train = death.head(101)
death_test = death.tail(27)

recovered_train = recovered.head(101)
recovered_test = recovered.tail(27)
```

```python
death_train_x,death_train_y = process_data(np.array(death_train))
death_test_x,death_test_y = process_data(np.array(death_test))

recovered_train_x,recovered_train_y = process_data(np.array(recovered_train))
recovered_test_x,recovered_test_y = process_data(np.array(recovered_test))
```

```python
semua_data = [
    [death_train_x,death_test_x,death_train_y,death_test_y,'Total Kematian'],
    [recovered_train_x,recovered_test_x,recovered_train_y,recovered_test_y,'Total Sembuh']
]
```

# RNN Model

```python
death_model = keras.Sequential([
    keras.layers.LSTM(8,input_shape=(1,1,)),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1)
])
```

```python
recovered_model = keras.Sequential([
    keras.layers.LSTM(8,input_shape=(1,1,)),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1)
])
```

```python
death_model.compile(loss='mse',optimizer='adam')
```

```python
recovered_model.compile(loss='mse',optimizer='adam')
```

# Create Callback & EarlyStopping

```python
os.makedirs('logs',exist_ok=True)
logdir = os.path.join('logs',datetime.datetime.now().strftime('%Y%m%d-%H%M%S'))
```

```python
callback = keras.callbacks.TensorBoard(logdir)
earlyStoping = keras.callbacks.EarlyStopping(monitor='loss',patience=3)
```

# Training Model

```python
death_history = death_model.fit(death_train_x,death_train_y,epochs=500,verbose=2,batch_size=32,callbacks=[callback,earlyStoping])
```

```python
recovered_history = recovered_model.fit(recovered_train_x,recovered_train_y,epochs=500,verbose=2,batch_size=32,callbacks=[callback,earlyStoping])
```

# Prediction

```python
death_prediction = death_model.predict(death_test_x)
recovered_prediction = recovered_model.predict(recovered_test_x)
```

```python
def visualisasiPrediksi(y_true,y_pred):
    plt.figure(figsize=(10,5))
    plt.title('Prediksi vs Aktual')
    plt.plot(y_true,'r',label='Aktual')
    plt.plot(y_pred,'b',label='Prediksi')
    plt.legend()
    plt.show()
```

**Prediksi vs Aktual**

Semakin dekat (hingga menumpuk) 2 grafik, maka semakin tepat prediksi model

**Selisih**

Semakin datar grafik (hingga tepat horizontal) maka semakin tepat prediksi model

```python
visualisasiPrediksi(death_test_y,death_prediction)
```

```python
visualisasiPrediksi(recovered_test_y,recovered_prediction)
```
